```python
import pandas as pd

df = pd.DataFrame({"cap-color": ["n", "b", "g", "r", "p", "u", "e", "w", "y", "l", "o", "k"]})
print("The original data")
print(df)
print("*" * 30)
df_new = pd.get_dummies(df, columns=["cap-color"], prefix="cap-color")
print("The transform data using get_dummies")
print(df_new)


```
